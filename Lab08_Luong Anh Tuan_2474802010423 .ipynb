```python
import nltk
# Tải trước các gói tài nguyên cần thiết từ NLTK
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('names')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
nltk.download('punkt_tab')
nltk.download('tagsets_json')
nltk.download('averaged_perceptron_tagger_eng')
sample_text = "The quick brown fox jumps over the lazy dog. Programming in Python is fun and educational."
```

```python
# =====================================================================
# Câu 1: Viết chương trình Python với thư viện NLTK để liệt kê các tên của corpus.
# =====================================================================
print("\n--- Câu 1: Danh sách các corpus có sẵn tải về (hoặc gói dữ liệu mẫu) ---")
import nltk.corpus
print("Một số tên gói dữ liệu/corpus phổ biến trong NLTK:")
corpus_names = [name for name in dir(nltk.corpus) if not name.startswith('_')]
print(corpus_names[:15])
```

```python
# =====================================================================
# Câu 2: Viết chương trình Python với thư viện NLTK để liệt kê danh sách các stopword bằng các ngôn ngữ khác nhau.
# =====================================================================
print("\n--- Câu 2: Liệt kê các ngôn ngữ hỗ trợ Stopwords ---")
from nltk.corpus import stopwords
languages = stopwords.fileids()
print("Các ngôn ngữ được hỗ trợ stopword trong NLTK:")
print(languages)
```

```python
# =====================================================================
# Câu 3: Viết chương trình Python với thư viện NLTK để kiểm tra danh sách các stopword bằng các ngôn ngữ khác nhau.
# =====================================================================
print("\n--- Câu 3: Kiểm tra danh sách stopword của một vài ngôn ngữ khác nhau ---")
print("English stopwords (10 từ đầu):", stopwords.words('english')[:10])
print("French stopwords (10 từ đầu):", stopwords.words('french')[:10])
print("German stopwords (10 từ đầu):", stopwords.words('german')[:10])
```

```python
# =====================================================================
# Câu 4: Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.
# =====================================================================
print("\n--- Câu 4: Loại bỏ các stopword từ văn bản ---")
from nltk.tokenize import word_tokenize

stop_words_en = set(stopwords.words('english'))
tokens = word_tokenize(sample_text)
filtered_tokens = [w for w in tokens if w.lower() not in stop_words_en]

print("Văn bản gốc:", sample_text)
print("Sau khi loại bỏ stopword:", filtered_tokens)
```

```python
# =====================================================================
# Câu 5: Viết chương trình Python với thư viện NLTK bỏ qua các stopword từ danh sách các stopword.
# =====================================================================
print("\n--- Câu 5: Loại bỏ một số từ chỉ định ra khỏi danh sách stopword mặc định ---")
words_to_keep = {'the', 'over', 'and'}
custom_stopwords = set(stopwords.words('english')) - words_to_keep

filtered_custom = [w for w in tokens if w.lower() not in custom_stopwords]
print("Mã sau khi giữ lại các từ 'the', 'over', 'and' từ stopword:")
print(filtered_custom)
```

```python
# =====================================================================
# Câu 6: Viết một chương trình Python với thư viện NLTK để tìm định nghĩa và ví dụ của một từ đã cho bằng WordNet từ Wikipedia.
# =====================================================================
print("\n--- Câu 6: Tìm định nghĩa và ví dụ của một từ bằng WordNet ---")
from nltk.corpus import wordnet as wn

word_to_search = "programming"
synsets = wn.synsets(word_to_search)

if synsets:
    print(f"Từ cần tra: '{word_to_search}'")
    print("Định nghĩa:", synsets[0].definition())
    print("Ví dụ:", synsets[0].examples())
else:
    print(f"Không tìm thấy dữ liệu WordNet cho từ '{word_to_search}'")
```

```python
# =====================================================================
# Câu 7: Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó.
# =====================================================================
print("\n--- Câu 7: Tìm tập hợp từ đồng nghĩa và trái nghĩa ---")
word_target = "good"
synonyms = set()
antonyms = set()

for syn in wn.synsets(word_target):
    for l in syn.lemmas():
        synonyms.add(l.name())
        if l.antonyms():
            antonyms.add(l.antonyms()[0].name())

print(f"Từ mục tiêu: '{word_target}'")
print("Từ đồng nghĩa (Synonyms):", list(synonyms)[:10])
print("Từ trái nghĩa (Antonyms):", list(antonyms)[:10])
```

```python
# =====================================================================
# Câu 8: Viết chương trình Python với thư viện NLTK để có cái nhìn tổng quan về bộ tag, chi tiết của một tag cụ thể trong bộ tag và chi tiết về một số bộ tag liên quan, sử dụng biểu thức chính quy.
# =====================================================================
print("\n--- Câu 8: Tổng quan về bộ tag (POS Tagset) ---")
print("Chi tiết về thẻ 'NN' (Noun):")
nltk.help.upenn_tagset('NN')

print("\nChi tiết về thẻ 'VB' (Verb):")
nltk.help.upenn_tagset('VB')
```

```python
# =====================================================================
# Câu 9: Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai danh từ đã cho.
# =====================================================================
print("\n--- Câu 9: So sánh độ tương đồng (similarity) giữa hai danh từ ---")
noun1 = wn.synset('dog.n.01')
noun2 = wn.synset('cat.n.01')

similarity_nouns = noun1.wup_similarity(noun2)
print(f"Độ tương đồng giữa danh từ '{noun1.name()}' và '{noun2.name()}':", similarity_nouns)
```

```python
# =====================================================================
# Câu 10: Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai động từ đã cho.
# =====================================================================
print("\n--- Câu 10: So sánh độ tương đồng giữa hai động từ ---")
verb1 = wn.synset('run.v.01')
verb2 = wn.synset('walk.v.01')

similarity_verbs = verb1.wup_similarity(verb2)
print(f"Độ tương đồng giữa động từ '{verb1.name()}' và '{verb2.name()}':", similarity_verbs)
```

```python
# =====================================================================
# Câu 11: Viết chương trình Python với thư viện NLTK để tìm số lượng tên nam và nữ trong các tên kho ngữ liệu. In tên 10 nam và nữ đầu tiên.
# =====================================================================
print("\n--- Câu 11: Thống kê và hiển thị tên nam, nữ từ kho dữ liệu 'names' ---")
from nltk.corpus import names

male_names = names.words('male.txt')
female_names = names.words('female.txt')

print("Tổng số lượng tên Nam (male.txt):", len(male_names))
print("Tổng số lượng tên Nữ (female.txt):", len(female_names))
print("10 tên Nam đầu tiên:", male_names[:10])
print("10 tên Nữ đầu tiên:", female_names[:10])
```

```python
# =====================================================================
# Câu 12: Viết chương trình Python với thư viện NLTK để in 15 kết hợp từ (Collocations/Bigrams) phổ biến nhất từ một kho văn bản.
# =====================================================================
print("\n--- Câu 12: In 15 kết hợp từ (Bigram Collocations) phổ biến ---")
from nltk.collocations import BigramAssocMeasures, BigramCollocationFinder

words_list = word_tokenize("The quick brown fox jumps over the lazy dog. The quick brown fox is very fast. A quick brown fox is smart.")

finder = BigramCollocationFinder.from_words(words_list)
measures = BigramAssocMeasures()

top_15_bigrams = finder.nbest(measures.raw_freq, 15)
print("15 kết hợp từ phổ biến nhất trong chuỗi thử nghiệm:")
print(top_15_bigrams)
```

```python
# =====================================================================
# Câu 13: Thực hiện gán nhãn từ loại (POS Tagging) cho một đoạn văn bản và đếm số lượng các loại từ loại khác nhau.
# =====================================================================
print("\n--- Câu 13: Gán nhãn từ loại (POS Tagging) và đếm tần số từ loại ---")
from collections import Counter

# Tiến hành phân tách từ và gán nhãn từ loại
text_tokens = word_tokenize(sample_text)
pos_tags = nltk.pos_tag(text_tokens)

print("Kết quả gán nhãn từ loại cho văn bản đoạn văn mẫu:")
print(pos_tags)

# Đếm tần suất xuất hiện của mỗi bộ nhãn từ loại
tag_counts = Counter(tag for word, tag in pos_tags)
print("\nTần số của các loại từ loại xuất hiện:")
for tag, count in tag_counts.items():
    print(f" - Từ loại [{tag}]: {count} lần")

print("\n--- KẾT THÚC LAB ---")
```
